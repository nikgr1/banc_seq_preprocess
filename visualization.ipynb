```python
data = pd.read_csv(path_to_file, sep='\t')
data.head()
```

```python
data['Length'] = data['End'] - data['Start']
data.head()
```

```python
# columns = ['fcOverControl', 'kd', 'p', 'r', 'n', 'Length']
# print(columns)
# for col in columns:
#     print(col)
#     plot = data[col].plot(kind='hist', edgecolor='black', logy=True, title='%s' % col)
#     plt.show()

columns = ['fcOverControl', 'kd', 'p', 'r', 'n', 'Length']

fig, axes = plt.subplots(tight_layout=True)
fig.set_size_inches(7, 14)
for i, col in enumerate(columns, 1):
    ax = plt.subplot(6, 2, i)
    # plot = data[col].plot(kind='hist', logy=True, title='%s' % col, xlim=(0, 1.2), bins=50, sharex=True, ax=ax) #edgecolor='black', 
    plot = data[col].plot(kind='hist', edgecolor='black', logy=True, title='%s' % col)
plt.show()
```

```python
import seaborn as sns

columns = ['1000_nM_YY1', '0100_nM_YY1', '0010_nM_YY1', '0001_nM_YY1', '0000_nM_YY1', ]
melted = pd.melt(data[columns])
melted = melted.rename(columns={'variable': 'TF concentration', 'value': 'Theta'})

ax = sns.boxplot(x="TF concentration", y="Theta", data=melted)
ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
plt.show()
```

```python
ax = sns.violinplot(x="TF concentration", y="Theta", data=melted)
ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
plt.show()
```

```python
melted['Log10 Theta'] = np.log10(melted['Theta'])

ax = sns.violinplot(x="TF concentration", y="Log10 Theta", data=melted)
ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
plt.show()
```
